```python
import pandas as pd
import numpy as np
```

```python
from pybroker import YFinance
yfinance = YFinance()
```

```python
# {'DBC': diversified index, 'GLD':gold, 'SLV':silver, 'PPLT':platinum, 'PALL':palladium, 'CPER':copper, 'DBB': base metals index, 'UNG':gas, 'USO':oil, 'UGN':gasoline, 'DBA': agriculture index, 'CORN':corn, }
commodity_list = ['DBC', 'IAU', 'SLV', 'PPLT', 'PALL', 'CPER', 'DBB', 'UNG', 'USO', 'UGN', 'DBA', 'CORN', 'WEAT', ]

# india: 'INDA', 'INDY'; vietnam: 'VNM'; china: 'FXI', 'GXC'; mexico: 'EWW'; japan: 'EWJ', 'DXJ'; taiwan: 'EWT'; korea:'EWY', 'KPOP', 'FLKR'
other_list = ['INDA', 'INDY', 'VNM', 'FXI', 'GXC', 'EWW', 'EWJ', 'DXJ', 'EWT', 'EWY', 'KPOP', 'FLKR', 'VXX', 'DIA', 'QQQ', 'SPY']
```

```python
# YFinance().query(spx_list[0:3], start_date='1/1/2017',end_date='15/1/2017', _timeframe='1d', _adjust='all')
df_toadd = yfinance.query(commodity_list+other_list, start_date='1/1/2017', end_date='15/5/2024')
df_toadd.head()
```

```python
df_spx500 = pd.read_csv("../data/processed/spx500_2017-2024_notnull.csv").drop(["Unnamed: 0.1", "Unnamed: 0"], axis=1)
df_spx500.head()
```

```python
# from datetime64 to string
df_toadd['date'] = df_toadd['date'].dt.strftime('%Y-%m-%d')
```

```python
df_toadd.date
```

```python
df_all = pd.concat([df_toadd, df_spx500], ignore_index=True)
df_all
```

```python
df_all.to_csv("../data/processed/spx500+commodity+emerging_2017-2024_notnull.csv", index=False)
```
