```python
# to build machine learning models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, recall_score
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
```

```python
features = ['volume', 'adj_close', 'hlc', 'coo', 'cll', 'chh', 'return', 'MA5', 'MA20', 'MA50', 'c-MA5', 'c-MA20', 'c-MA50', 'EMA_12', 'EMA_26', 'MACD', 'Signal_Line', 'MACD_Buy_Signal', 'MACD_Sell_Signal', 'BB_Middle', 'BB_Upper', 'BB_Lower', 'BB_Buy_Signal', 'BB_Sell_Signal', 'RSI_14']
target = 'Y5'
target_int = 'Y5_int'
```

```python
cluster_num = 4
cluster_df = df[df.cluster == cluster_num].copy()
train_end_date = '2021-01-01'

train_data = cluster_df[cluster_df['date'] < train_end_date]
test_data = cluster_df[cluster_df['date'] >= train_end_date]
X_train = train_data[features]
y_train = train_data[target]
y_train_int = train_data[target_int]
X_test = test_data[features]
y_test = test_data[target]
y_test_int = test_data[target_int]
```

```python
# Standard Scaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
models = {
    'Lasso': Lasso(),
    'RandomForest': RandomForestRegressor(),
    'AdaBoost': AdaBoostRegressor(),
}

for model_name, model in models.items():
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    y_train_pred_int = pd.Series(y_train_pred).apply(lambda x: 1 if x > 0 else -1)
    y_test_pred = model.predict(X_test)
    y_test_pred_int = pd.Series(y_test_pred).apply(lambda x: 1 if x > 0 else -1)
    
    train_accuracy = accuracy_score(y_train_int, y_train_pred_int)
    train_recall = recall_score(y_train_int, y_train_pred_int, pos_label=1)
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_r2 = r2_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test_int, y_test_pred_int)
    test_recall = recall_score(y_test_int, y_test_pred_int, pos_label=1)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    results_list.append({
        'Cluster': cluster_num,
        'Model': model_name,
        'Train_Accuracy': train_accuracy,
        'Train_Recall': train_recall,
        'Train_MSE': train_mse,
        'Train_R2': train_r2,
        'Test_Accuracy': test_accuracy,
        'Test_Recall': test_recall,
        'Test_MSE': test_mse,
        'Test_R2': test_r2
    })
    
    print(cluster_num, " ", model_name," finished")
    joblib.dump(model, f'{model_name}_model_cluster_{cluster_num}.pkl')
```

```python
def train_model(model, cluster_num = 0, target='Y5', train_end_date = '2021-01-01'):
    cluster_df = df[df.cluster == cluster_num].copy()
    target_int = target+'_int'
    
    train_data = cluster_df[cluster_df['date'] < train_end_date]
    test_data = cluster_df[cluster_df['date'] >= train_end_date]
    X_train = train_data[features]
    y_train = train_data[target]
    y_train_int = train_data[target_int]
    X_test = test_data[features]
    y_test = test_data[target]
    y_test_int = test_data[target_int]
    
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    print('model training prepared: ')
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    y_train_pred_int = pd.Series(y_train_pred).apply(lambda x: 1 if x > 0 else -1)
    y_test_pred = model.predict(X_test)
    y_test_pred_int = pd.Series(y_test_pred).apply(lambda x: 1 if x > 0 else -1)
    print('test output: ', 'model:', model_name, '\t', y_test_pred[:20], y_test_pred_int.values[:20])
    
    # combinded to df_signal
    
    train_accuracy = accuracy_score(y_train_int, y_train_pred_int)
    train_recall = recall_score(y_train_int, y_train_pred_int, pos_label=1)
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_r2 = r2_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test_int, y_test_pred_int)
    test_recall = recall_score(y_test_int, y_test_pred_int, pos_label=1)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    results_list.append({
        'Cluster': cluster_num,
        'Model': model_name,
        'Target': target,
        'Train_Accuracy': train_accuracy,
        'Train_Recall': train_recall,
        'Train_MSE': train_mse,
        'Train_R2': train_r2,
        'Test_Accuracy': test_accuracy,
        'Test_Recall': test_recall,
        'Test_MSE': test_mse,
        'Test_R2': test_r2
    })
    
    print(cluster_num, " ", model_name," finished")
    joblib.dump(model, f'{model_name}_model_cluster_{cluster_num}_{target}.pkl')
```
